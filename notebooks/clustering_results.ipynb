```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("../data/final_cocktail_dataset_with_clusters.parquet")
df.head(n=2).T
```

```python
df["clusters"].unique()
```

```python
# Set up the plot
plt.figure(figsize=(10, 7))

# Create scatter plot with cluster coloring
scatter = plt.scatter(df['UMAP1'], df['UMAP2'], c=df['clusters'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title("UMAP Projection of Cocktails with KMeans Clusters")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```
