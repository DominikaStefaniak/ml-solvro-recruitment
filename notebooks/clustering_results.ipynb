# Qualitative evaluation of clustering models

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("../data/final_cocktail_dataset_with_clusters.parquet")
df.head(n=2).T
```

```python
df["KMeans_clusters"].unique()
```

```python
plt.figure(figsize=(10, 7))

scatter = plt.scatter(df['UMAP1'], df['UMAP2'], c=df['KMeans_clusters'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title("UMAP Projection of Cocktails with KMeans Clusters")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```

```python
plt.figure(figsize=(10, 7))

scatter = plt.scatter(df['UMAP1'], df['UMAP2'], c=df['DBSCAN_clusters'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title("UMAP Projection of Cocktails with DBSCAN Clusters")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```

# Summary
Both clustering methods separated the data into 6 clusters.  
  
1. **Quantitative evaluation** results as a reminder:  
**Silhouette score:** 0.8276004195213318, **Davies Bouldin score:** 0.2234392830430084.  
These scores suggest that the clusters are well-separated from each other (since the Silhouette score is close to 1), and each cluster has low variation within it compared to the differences between clusters (which is indicated by a lower Davies-Bouldin score).  

2. Conclusions for **qualitative evaluation**:  
Even though visually one of the clusters looks like it should be devided into two separate clusters, according to the quantitative evaluation metrics the 6 clusters that both of the models created is best. This might be caused by the information loss while reducing the dimensionality from 10D to 2D, since the models and quantitative evaluation was performed on 10D dataset.
